#### Import des données

```python
# ------------------ #
# IMPORT DES DONNÉES #
# ------------------ #


def import_df(df_name, path_data, sep, index_col=None):
    df[df_name] = pd.read_csv(path_data+df_name+'.csv', sep = sep, index_col=index_col, low_memory=False).replace({'\r\n': '\n'}, regex=True)

def import_dfs(df_names, path_data, sep = ',', index_col=None, verbose=False):
    for df_name in tqdm(df_names) : 
        if(verbose) :
            print(" - ", df_name)
        import_df(df_name, path_data, sep, index_col=index_col)

tables_with_id = [
    'intrant', 
    'semence'
]

tables_without_id = [
]

# import des données de l'entrepôt avec la colonne 'id' en index 
import_dfs(tables_with_id, ENTREPOT_PATH, sep = ',', index_col='id', verbose=False)

# import des données du magasin
import_dfs(tables_without_id, ENTREPOT_PATH, sep = ',', verbose=False)
```

```python
# on ne sélectionne que les traitements de semences
df['traitement_semence'] = df['intrant'].loc[
    df['intrant']['type'] == 'TRAITEMENT_SEMENCE'
].reset_index()
```

```python
# on ajoute au traitement de semence l'information du lot dans lequel il est et du composant de culture sur lequel il porte
left = df['traitement_semence']
right = df['semence'][['lot_semence_id', 'composant_culture_id']]
df['traitement_semence_extanded'] = pd.merge(left, right, left_on='semence_id', right_index=True, how='left')
```

```python
# on groupe par lot_semenbce_id et ref_nom pour obtenir l'occurrence d'un même traitement dans un lot de semence
df['meme_traitement_dans_lot'] = df['traitement_semence_extanded'].groupby(['lot_semence_id', 'composant_culture_id', 'ref_id']).agg({
    'id' : 'count'
}).reset_index().rename(columns={'id' : 'nombre'})
```

```python
# on ne garde que ceux qui sont à priori problématiques
df['lot_semence_problematiques'] = df['meme_traitement_dans_lot'].loc[
    df['meme_traitement_dans_lot']['nombre'] > 1
]
```

```python
# on rajoute les information contextuelles pour vérifier sur l'interface : 
left = df['lot_semence_problematiques']
right = df['intrant'][['domaine_id', 'nom_utilisateur']]
merge = pd.merge(left, right, left_on='lot_semence_id', right_index=True, how='left')
```
