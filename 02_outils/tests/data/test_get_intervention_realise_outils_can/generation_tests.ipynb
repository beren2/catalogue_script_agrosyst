```python
df = {}

def import_df(df_name, path_data, sep, index_col=None):
    df[df_name] = pd.read_csv(path_data+df_name+'.csv', sep = sep, index_col=index_col, low_memory=False)

def import_dfs(df_names, path_data, sep = ',', index_col=None, verbose=False):
    for df_name in tqdm(df_names) : 
        if(verbose) :
            print(" - ", df_name)
        import_df(df_name, path_data, sep, index_col=index_col)

# get_intervention_realise_outils_can_context --> intervention_realise
# get_intervention_realise_action_outils_can --> 'intervention_realise', 'action_realise'
# get_intervention_realise_combinaison_outils_can -->  'intervention_realise', 'intervention_realise_restructure', 'combinaison_outil', 'materiel', 'combinaison_outil_materiel'
# get_intervention_realise_culture_outils_can --> 'intervention_realise', 'noeuds_realise', 'connection_realise', 
        # 'plantation_perenne_phases_realise', 
        # 'plantation_perenne_realise', 'composant_culture_concerne_intervention_realise', 
        # 'noeuds_realise_restructure', 'plantation_perenne_realise_restructure',
        # 'ccc_intervention_realise_restructure', 'composant_culture',
        # 'espece', 'variete', 'connection_realise_restructure'
# get_intervention_realise_culture_prec_outils_can --> 'composant_culture', 'espece', 'variete', 'intervention_realise', 
        # 'noeuds_realise', 'connection_realise', 'culture', 
        # 'noeuds_realise_restructure'

tables_entrepot = [
        'intervention_realise', 'action_realise', 'combinaison_outil', 'materiel',
        'noeuds_realise', 'connection_realise', 'plantation_perenne_phases_realise', 'plantation_perenne_realise', 'composant_culture_concerne_intervention_realise',
        'composant_culture', 'espece', 'variete', 'culture', 'intervention_realise_agrege', 'dispositif',
        'semence', 'utilisation_intrant_realise', 'intrant', 'recolte_rendement_prix', 'nuisible_edi', 'adventice',
        'nuisible_edi', 'utilisation_intrant_cible'
]

tables_performance = [
    'combinaison_outil_materiel'
]

# import des données de l'entrepôt avec la colonne 'id' en index 
import_dfs(tables_entrepot, ENTREPOT_PATH, sep = ',',index_col='id',verbose=False) 
import_dfs(tables_performance, ENTREPOT_PATH, sep = ',',verbose=False) 
```

```python
df['intervention_realise_test'] = df['intervention_realise'].loc[
    df['intervention_realise'].index.isin(studied_ids)
]
df['action_realise_test'] = df['action_realise'].loc[
    df['action_realise']['intervention_realise_id'].isin(df['intervention_realise_test'].index)
]


df['combinaison_outil_test'] = df['combinaison_outil'].loc[
    df['combinaison_outil'].index.isin(df['intervention_realise_test']['combinaison_outil_id'])
]
df['combinaison_outil_materiel_test'] = df['combinaison_outil_materiel'].loc[
    df['combinaison_outil_materiel']['combinaison_outil_id'].isin(df['combinaison_outil_test'].index)
]
df['materiel_test'] = df['materiel'].loc[
    df['materiel'].index.isin(df['combinaison_outil_materiel_test']['materiel_id'])
]

df['noeuds_realise_test'] = df['noeuds_realise'].loc[
    (df['noeuds_realise'].index.isin(df['intervention_realise_test']['noeuds_realise_id']))
]

df['connection_realise_test'] = df['connection_realise'].loc[
    (df['connection_realise']['cible_noeuds_realise_id'].isin(df['noeuds_realise_test'].index)) | 
    (df['connection_realise']['source_noeuds_realise_id'].isin(df['noeuds_realise_test'].index)) 
]

df['plantation_perenne_phases_realise_test'] = df['plantation_perenne_phases_realise'].loc[
    df['plantation_perenne_phases_realise'].index.isin(df['intervention_realise_test']['plantation_perenne_phases_realise_id'])
]
df['plantation_perenne_realise_test'] = df['plantation_perenne_realise'].loc[
    df['plantation_perenne_realise'].index.isin(df['plantation_perenne_phases_realise_test']['plantation_perenne_realise_id'])
]
df['composant_culture_concerne_intervention_realise_test'] = df['composant_culture_concerne_intervention_realise'].loc[
    df['composant_culture_concerne_intervention_realise']['intervention_realise_id'].isin(df['intervention_realise_test'].index)
]

df['culture_test'] = df['culture'].loc[
    df['culture'].index.isin(df['noeuds_realise_test']['culture_id'])
]
df['composant_culture_test'] = df['composant_culture'].loc[
    df['composant_culture']['culture_id'].isin(list(df['noeuds_realise_test']['culture_id'])) |
    df['composant_culture']['culture_id'].isin(list(df['plantation_perenne_realise_test']['culture_id'])) |
    df['composant_culture']['culture_id'].isin(list(df['connection_realise_test']['culture_intermediaire_id']))
]

df['espece_test'] = df['espece'].loc[
    df['espece'].index.isin(df['composant_culture_test']['espece_id'])
]
df['variete_test'] = df['variete'].loc[
    df['variete'].index.isin(df['composant_culture_test']['variete_id'])
]

df['intervention_realise_agrege_test'] = df['intervention_realise_agrege'].loc[
    df['intervention_realise_agrege'].index.isin(studied_ids)
]

df['dispositif_test'] = df['dispositif'].loc[
    df['dispositif'].index.isin(df['intervention_realise_agrege_test']['dispositif_id'])
]
df['semence_test'] = df['semence'].loc[
    df['semence']['composant_culture_id'].isin(df['composant_culture_test'].index)
]
df['utilisation_intrant_realise_test'] = df['utilisation_intrant_realise'].loc[
    df['utilisation_intrant_realise']['intervention_realise_id'].isin(df['intervention_realise_test'].index)
]
df['intrant_test'] = df['intrant'].loc[
    df['intrant'].index.isin(df['utilisation_intrant_realise_test']['intrant_id'])
]


df['recolte_rendement_prix_test'] = df['recolte_rendement_prix'].loc[
    df['recolte_rendement_prix']['action_id'].isin(df['action_realise_test'].index)
]

df['utilisation_intrant_cible_test'] = df['utilisation_intrant_cible'].loc[
    df['utilisation_intrant_cible']['utilisation_intrant_id'].isin(df['utilisation_intrant_realise_test'].index)
]
df['nuisible_edi_test'] = df['nuisible_edi'].loc[
    df['nuisible_edi'].index.isin(df['utilisation_intrant_cible_test']['ref_cible_id'])
]
df['adventice_test'] = df['adventice'].loc[
    df['adventice'].index.isin(df['utilisation_intrant_cible_test']['ref_cible_id'])
]
```
