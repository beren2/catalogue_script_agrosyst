```python
def import_df(df_name, path_data, sep, df):
    """
        importe un dataframe au chemin path_data+df_name+'.csv' et le stock dans le dictionnaire 'df' à la clé df_name
    """
    df[df_name] = pd.read_csv(path_data+df_name+'.csv', sep = sep).replace({'\r\n': '\n'}, regex=True)

def import_dfs(df_names, path_data,  df, sep = ','):
    """
        stocke dans le dictionnaire df tous les dataframes indiqués dans la liste df_names
    """
    for df_name in df_names : 
        import_df(df_name, path_data, sep, df)
    return df

df_names =  [   'sdc',
                'parcelle',
                'zone',
                'synthetise',
                'dispositif',
                'domaine'
            ]


donnees = import_dfs(df_names, ENTREPOT_PATH, {}, sep = ',')
```

```python
# dict : 
# cle : nom de la table a filtrer
# value : liste 0 : 'col_interne' si la colonne qui permet de filtrer est au sein du data à filtrer SINON 'col_externe' : 
#           liste 1 : table qui permet de filtrer , peut etre une liste de tables
#           liste 2 : colonne qui permet de filtrer de element 0

# ex1 : 'domaine' : ['col_externe','parcelle','domaine_id']
# ex2 : 'noeuds_realise' : ['col_interne','zone','zone_id']
# ex3 avec plusieurs tables qui permettent de filtrer: 'sdc' : ['col_externe',['zone','synthetise'],'sdc_id']

lien_entre_table = {'parcelle' : ['col_externe','zone','parcelle_id'],
                    'sdc' : ['col_externe',['parcelle','synthetise'],'sdc_id'],
                    'dispositif' : ['col_externe','sdc','dispositif_id'],
                    'domaine' : ['col_externe','dispositif','domaine_id']
}

data_filtre = donnees.copy()
data_filtre['zone'] = data_filtre['zone'].loc[data_filtre['zone']['id'].isin(studied_ids)]
data_filtre['synthetise'] = data_filtre['synthetise'].loc[data_filtre['synthetise']['id'].isin(studied_ids)]

print(data_filtre['zone'].shape)
print(data_filtre['synthetise'].shape)

def filtre_table(type_filtre,data_filtre,table_a_filtrer,table_filtre,colonne_pour_filtrer):
    if type_filtre == 'col_interne':
        res = data_filtre[table_a_filtrer].loc[data_filtre[table_a_filtrer][colonne_pour_filtrer].isin(data_filtre[table_filtre]['id'])]
    else :
        res = data_filtre[table_a_filtrer].loc[data_filtre[table_a_filtrer]['id'].isin(data_filtre[table_filtre][colonne_pour_filtrer])]
    return(res)

# parcours du dict des liens entre table pour filtrer progressivement les donnees lies aux id choisis
for key in lien_entre_table.keys() :
    table_a_filtrer = key
    print(table_a_filtrer)
    type_filtre = lien_entre_table[key][0]
    table_filtre = lien_entre_table[key][1]
    colonne_pour_filtrer = lien_entre_table[key][2]

    print(data_filtre[table_a_filtrer].shape)
    if type(table_filtre) == list: # cas avec plusieurs tables filtres
        temp = filtre_table(type_filtre,data_filtre,table_a_filtrer,table_filtre[0],colonne_pour_filtrer)
        for t in table_filtre[1:len(table_filtre)] :
            temp = pd.concat([temp, filtre_table(type_filtre,data_filtre,table_a_filtrer,t,colonne_pour_filtrer)])
        data_filtre[table_a_filtrer] = temp
    else : 
        data_filtre[table_a_filtrer] = filtre_table(type_filtre,data_filtre,table_a_filtrer,table_filtre,colonne_pour_filtrer)

    print(data_filtre[table_a_filtrer].shape)
```
