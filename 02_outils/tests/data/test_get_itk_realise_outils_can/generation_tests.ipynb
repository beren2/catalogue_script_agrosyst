```python
def get_synthetise_containing_only_one_perenne(searching_str = 'Poirier'):
    # on commence par sélectionner tous les noeuds qui nous intéresse : 
    left = df['plantation_perenne_synthetise']
    right = df['plantation_perenne_synthetise_restructure']
    df['plantation_perenne_synthetise_extanded'] = pd.merge(left, right, left_index=True, right_index=True, how='left')


    # on sélectionne uniquement les pommiers
    left = df['culture']
    right = df['composant_culture']
    merge = pd.merge(left, right, left_index=True, right_on='culture_id', how='left')

    # on ajoute l'espèce au composant de culture
    left = df['composant_culture']
    right = df['espece'][['libelle_espece_botanique']]
    df['composant_culture_extanded'] = pd.merge(left, right, left_on='espece_id', right_index=True, how='left')


    df['composant_culture_extanded']['contains_pomme'] = df['composant_culture_extanded']['libelle_espece_botanique'].str.contains(searching_str).values


    # on dit pour chaque culture si elle ne contient QUE des composant de culture "pomme"
    left = df['culture']
    right = df['composant_culture_extanded'].groupby('culture_id').agg({
        'contains_pomme' : "min"
    })

    df['culture_extanded'] = pd.merge(left, right, left_index=True, right_on='culture_id', how='left')

    df['culture_only_pomme'] = df['culture_extanded'].loc[
        df['culture_extanded']['contains_pomme'] == True
    ]

    # maintenant qu'on a toutes les cultures qui ne mobilisent QUE de la pomme, on peut sélectionner les perennes concernés : 
    df['plantation_perenne_pomme'] = df['plantation_perenne_synthetise_extanded'].loc[
        df['plantation_perenne_synthetise_extanded']['culture_id'].isin(df['culture_only_pomme'].index)
    ]

    list_plantation_only_pomme = list(df['plantation_perenne_synthetise'].loc[
        df['plantation_perenne_synthetise'].index.isin(df['plantation_perenne_pomme'].index)
    ].index)


    df['plantation_perenne_synthetise'].loc[:, 'is_only_pomme'] = False
    df['plantation_perenne_synthetise'].loc[list_plantation_only_pomme, 'is_only_pomme'] = True

    left = df['synthetise'].reset_index()
    right = df['plantation_perenne_synthetise'].groupby('synthetise_id').agg({
        'is_only_pomme' : "min"
    })
    df['synthetise_extanded'] = pd.merge(left, right, left_on='id', right_on='synthetise_id', how='left').set_index('id')
    
    return list(
        df['synthetise_extanded'].loc[ 
            df['synthetise_extanded']['is_only_pomme'] == True
        ].index
    )
```

```python
list_poirier = get_synthetise_containing_only_one_perenne(searching_str='Poirier')
list_pommier = get_synthetise_containing_only_one_perenne(searching_str='Pommier')
list_prunier = get_synthetise_containing_only_one_perenne(searching_str='Prunier')
```

```python
data = [(index, 'Poirier') for index in list_poirier] + \
       [(index, 'Pommier') for index in list_pommier] + \
       [(index, 'Prunier') for index in list_prunier]

# Convert the list of tuples to a DataFrame
res = pd.DataFrame(data, columns=['synthetise_id', 'libelle_espece_botanique'])
```
